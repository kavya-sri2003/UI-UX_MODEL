```python
import os
import pandas as pd
from PIL import Image
from datetime import datetime

# Path to your folder containing images
folder_path = 'UIUX_MODEL'

# List to store image file information
image_info = []

# Traverse the directory
for root, dirs, files in os.walk(folder_path):
    print(f"Scanning directory: {root}")  # Debug: Print the current directory being scanned
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            file_path = os.path.join(root, file)
            print(f"Processing file: {file_path}")  # Debug: Print the file being processed
            try:
                file_name = file
                file_size = os.path.getsize(file_path) / 1024  # size in KB
                file_extension = os.path.splitext(file)[1].lower()

                # Get image dimensions
                with Image.open(file_path) as img:
                    width, height = img.size

                # Get file creation and modification dates
                creation_time = os.path.getctime(file_path)
                modification_time = os.path.getmtime(file_path)
                creation_date = datetime.fromtimestamp(creation_time).strftime('%Y-%m-%d %H:%M:%S')
                modification_date = datetime.fromtimestamp(modification_time).strftime('%Y-%m-%d %H:%M:%S')

                # Append the image information
                image_info.append([file_path, file_name, file_size, file_extension, width, height, creation_date, modification_date])
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")  # Debug: Print errors if any

# Check if any data was collected
if not image_info:
    print("No image files were found or processed.")

# Create a DataFrame
df = pd.DataFrame(image_info, columns=['Image_Path', 'File_Name', 'File_Size (KB)', 'File_Extension', 'Width (px)', 'Height (px)', 'Creation_Date', 'Modification_Date'])

# Save DataFrame to CSV
df.to_csv('image_files_info.csv', index=False)

print("CSV file created successfully!")
```
